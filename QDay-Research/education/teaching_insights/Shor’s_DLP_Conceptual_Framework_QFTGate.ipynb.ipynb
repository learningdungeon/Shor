```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.circuit.library import QFTGate  # Use QFTGate

def shors_dlog_fixed(n_qubits):
    """
    QFTGate
    """
    qc = QuantumCircuit(2*n_qubits, n_qubits)
    
    # Apply Hadamard gates to the control register
    for qubit in range(n_qubits):
        qc.h(qubit)
    
    # Placeholder for elliptic curve operations
    for qubit in range(n_qubits):
        qc.cx(qubit, n_qubits + qubit)
    
    # Apply QFT using QFTGate - NEW WAY
    qft_gate = QFTGate(num_qubits=n_qubits)
    qc.append(qft_gate, range(n_qubits))
    
    qc.measure(range(n_qubits), range(n_qubits))
    return qc

# Test it
n = 4
qc = shors_dlog_fixed(n)
print("Circuit with QFTGate:")
print(qc.draw())
```
